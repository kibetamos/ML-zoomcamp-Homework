```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/home/bench/Downloads/2009.csv')

# Check basic information about the dataset
print(df.info())
print(df.head())

# Display summary statistics
print(df.describe())

# Check for missing values
print(df.isnull().sum())
```

```python
# Drop irrelevant columns (adjust based on your dataset)
df_cleaned = df.drop(['FL_DATE', 'CANCELLED', 'UNIQUE_CARRIER'], axis=1, errors='ignore')

```

```python
# Handle missing values (impute with mean for numerical columns)
df_cleaned.fillna(df_cleaned.mean(), inplace=True)

# One-hot encode categorical variables
df_encoded = pd.get_dummies(df_cleaned, drop_first=True)

```

```python
# Separate features and target
X = df_encoded.drop('DEP_DELAY', axis=1, errors='ignore')
y = df_encoded['DEP_DELAY']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Use 10% of the data for prototyping
X_sample, _, y_sample, _ = train_test_split(X, y, test_size=0.9, random_state=42)

# Train a lightweight Linear Regression model
model = LinearRegression()
model.fit(X_sample, y_sample)

# Print a simple result
print(f"Model coefficients: {model.coef_}")
```
